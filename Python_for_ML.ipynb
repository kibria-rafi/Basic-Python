# **4th LAB**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
pd.read_csv('cse.csv')
```

```python
dataSet = pd.read_csv('cse.csv',usecols=['Age','Cabin','Embarked'])
```

```python
dataSet
```

```python
def input_nan(df,var,mean):
  df[var+"_mean"]=df[var].fillna(mean)
```

```python
mean=dataSet["Age"].mean()
```

```python
input_nan(dataSet,"Age",mean)
```

```python
dataSet.isnull().sum()
```

```python
dataSet["Age"].plot(kind="kde")
dataSet["Age_mean"].plot(kind="kde")
```

```python
def input_nan1(df,var,median):
  df[var+"_median"]=df[var].fillna(mean)
```

```python
median=dataSet["Age"].median()
```

```python
input_nan1(dataSet,"Age",median)
```

```python
dataSet.isnull().sum()
```

```python
dataSet["Age"].plot(kind="kde")
dataSet["Age_median"].plot(kind="kde")
```

```python
def input_nan2(df,var,mode):
  df[var+"_mode"]=df[var].fillna(mode)
```

```python
mode=dataSet["Age"].mode()
```

```python
input_nan2(dataSet,"Age",mode)
```

```python
dataSet.isnull().sum()
```

```python
dataSet["Age"].plot(kind="kde")
dataSet["Age_mode"].plot(kind="kde")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
pd.read_csv('cse.csv')
```

```python
dataSet = pd.read_csv('cse.csv',usecols=['Age','Cabin','Embarked'])
```

```python
dataSet.isnull().sum()
```

```python
dataSet["Age"].dropna().sample(dataSet["Age"].isnull().sum(),random_state=0)
```

```python
dataSet["Age"].isnull().index
```

```python
def input_sample(df,var):
  df[var+"_sample"]=df[var]
  random_sample=dataSet[var].dropna().sample(dataSet[var].isnull().sum(),random_state=0)
  random_sample.index=df[df[var].isnull()].index
  df.loc[df[var].isnull(),var+"_sample"]=random_sample

```

```python
input_sample(dataSet,"Age")
```

```python
dataSet
```

```python
dataSet.isnull().sum()
```

```python
dataSet["Age"].plot(kind="kde")
dataSet["Age_sample"].plot(kind="kde")
```
